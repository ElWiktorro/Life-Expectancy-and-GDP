# Importing Modules And Loading Data 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('all_data.csv')

#showing few first records of dataframe
print(df.head())

#checking which countries' data are in dataframe 
print(df.Country.unique())
```

# Exploring and Explaing Data

# 1. Has life expectancy increased over time in the six nations?

```python
#splliting df dataframe for dataframes for each country
chile_df = df[df.Country == 'Chile']
china_df = df[df.Country == 'China']
germany_df = df[df.Country == 'Germany']
mexico_df = df[df.Country == 'Mexico']
usa_df = df[df.Country == 'United States of America']
zimbabwe_df = df[df.Country == 'Zimbabwe']

#presenting line charts for each country (Life expectancy at birth (years) vs Year) 
plt.figure(figsize=(20,10))
plt.subplot(2,3,1)
plt.plot(chile_df.Year, chile_df['Life expectancy at birth (years)'])
plt.title('Chile')
plt.ylabel('Life expectancy at birth (years)')
plt.subplot(2,3,2)
plt.plot(china_df.Year, china_df['Life expectancy at birth (years)'], color = 'orange')
plt.title('China')
plt.subplot(2,3,3)
plt.plot(germany_df.Year, germany_df['Life expectancy at birth (years)'], color = 'green')
plt.title('Germany')
plt.subplot(2,3,4)
plt.plot(mexico_df.Year, mexico_df['Life expectancy at birth (years)'], color = 'red')
plt.title('Mexico')
plt.ylabel('Life expectancy at birth (years)')
plt.subplot(2,3,5)
plt.plot(usa_df.Year, usa_df['Life expectancy at birth (years)'], color = 'purple')
plt.title('United States of America')
plt.subplot(2,3,6)
plt.plot(zimbabwe_df.Year, zimbabwe_df['Life expectancy at birth (years)'], color = 'brown')
plt.title('Zimbabwe')
plt.show()
```

The charts above show that Life expectancy increased over these 15 years.

# 2. Has GDP increased over time in the six nations?

```python
plt.figure(figsize = (20,10))
sns.lineplot(x=df.Year, y=df.GDP, hue = df.Country)
plt.ylabel('GDP in Trillions of U.S. Dollars')
plt.show()
plt.clf()

plt.figure(figsize=(20,10))
plt.subplot(2,3,1)
plt.plot(chile_df.Year, chile_df.GDP)
plt.title('Chile')
plt.ylabel('GDP in Trillions of U.S. Dollars')
plt.subplot(2,3,2)
plt.plot(china_df.Year, china_df.GDP, color = 'orange')
plt.title('China')
plt.subplot(2,3,3)
plt.plot(germany_df.Year, germany_df.GDP, color = 'green')
plt.title('Germany')
plt.subplot(2,3,4)
plt.plot(mexico_df.Year, mexico_df.GDP, color = 'red')
plt.title('Mexico')
plt.ylabel('GDP in Trillions of U.S. Dollars')
plt.subplot(2,3,5)
plt.plot(usa_df.Year, usa_df.GDP, color = 'purple')
plt.title('United States of America')
plt.subplot(2,3,6)
plt.plot(zimbabwe_df.Year, zimbabwe_df.GDP, color = 'brown')
plt.title('Zimbabwe')
plt.show()
plt.clf()
```

Looking at first chart we can say that GDP in China, USA and Germany has increased, but in other countries was the same over these years.

But when we looking at charts for each country seperately we can notice that GDP has increased in all of presented 6 countries.

# 3. Is there a correlation between GDP and life expectancy of a country?

```python
plt.figure(figsize=(20,10))
plt.subplot(2,3,1)
plt.scatter(chile_df['Life expectancy at birth (years)'], chile_df.GDP)
plt.title('Chile')
plt.ylabel('GDP in Trillions of U.S. Dollars')
plt.subplot(2,3,2)
plt.scatter(china_df['Life expectancy at birth (years)'], china_df.GDP, color = 'orange')
plt.title('China')
plt.subplot(2,3,3)
plt.scatter(germany_df['Life expectancy at birth (years)'], germany_df.GDP, color = 'green')
plt.title('Germany')
plt.subplot(2,3,4)
plt.scatter(mexico_df['Life expectancy at birth (years)'], mexico_df.GDP, color = 'red')
plt.title('Mexico')
plt.ylabel('GDP in Trillions of U.S. Dollars')
plt.subplot(2,3,5)
plt.scatter(usa_df['Life expectancy at birth (years)'], usa_df.GDP, color = 'purple')
plt.title('United States of America')
plt.subplot(2,3,6)
plt.scatter(zimbabwe_df['Life expectancy at birth (years)'], zimbabwe_df.GDP, color = 'brown')
plt.title('Zimbabwe')
plt.show()
```

Looking at above plots we can notice a positive correlation between GDP and Life expectancy

# 4. What is the average life expectancy in these nations?

```python
average_life_expectancy = df.groupby('Country')['Life expectancy at birth (years)'].mean().reset_index()
print(average_life_expectancy)
```
